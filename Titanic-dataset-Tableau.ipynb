# Summary

The data set contains information about 891 passengers out of 2224 passengers on board the Titanic. The dataset has 12 fields. The main fields of interest in this analysis are passenger class, age, sex and fare. Some fields have been calculated. Ex. Age group that splits the data into 3 groups based on age (Adult, child and Senior).

# Design


I wanted to create some basic plots to understand the structure of data. I began with a histogram of age and wanted answer two questions -

1) What did the distribution look like?

2) What was the age range that contained maximum number of survivors?

I decided to encode passenger survival of 1 or 0 to a yes or no so that all the rows are included in the analysis classified by their respective counts. When I used a 1 and 0, I noticed that some the agrregations were either left out or incorrect because the value 0 was affecting the result. 

Next, I created bar plots since I had sex and passenger class as categorical variables and I wanted to know who had a better survival rate? - 

1) males or females ? 

2) 1st class or 3rd class? 

**Color coding**

I decided to use color coding for mainly survival category being Yes/No because ultimately my analysis was based on understanding the factors affecting survival so I wanted that distinction to pop out. I also color coded the count in each category of the age histogram in order to highlight the bins with the most survivors.

**choice of charts**

The reason I used bar charts in most of the cases was also based on feedback I received because I was looking at survival counts for each category and across categories. I have also however used pie charts in one of the cases where one variable was under study and a represnetation of it pertaining to the whole mattered. Ex. survival by class. I have also marked the labels as percentages in this case in addition to raw numbers to give the user a better grasp of understanding of the data distribution along a whole pie.

**Interactions/Animations**
Some of the interactions I included are in the dashboards where I've provided some filter options and some hover options to select/deselect certain views in order for the user to drill down to specifics.
